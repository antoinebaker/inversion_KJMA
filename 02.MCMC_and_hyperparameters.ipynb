### MCMC sampling

The MAP estimate yields a very accurate reconstruction of the replication program, however it does not give information on the reliability (or posterior uncertaincy) of the estimate.

In order to estimate the posterior uncertainty, we sample the posterior distribution $P(I|d, \beta)$ using a MCMC method called the [Hamiltonian Monte Carlo algorithm](http://en.wikipedia.org/wiki/Hamiltonian_Monte_Carlo).

```python
# All computation will be done in the infer object from the ki.InferKJMA class
infer = ki.InferKJMA(x=x, t=t,
                     data=data, xdata=xdata, tdata=tdata,
                     v=v, sigma_d=sigma_d,
                     I_0=I_0, sigma_0=sigma_0, tau_0=tau_0, ell_0=ell_0)

# takes about 3 min to generate 500 MCMC samples 
# the auto-correlation length of the MCMC is about 10 samples (estimated by the ACF)
# so we get about 100 effectively independent samples
infer.compute_MCMC_samples(verbose=True)
```

Ten MCMC samples of the posterior distribution, for $T(x)$, $p(x)$, $\rho_{ini}(x)$ and $\rho_{ter}(x)$,
are shown by the blue lines below. We see that the posterior uncertainty is small.

```python
plot_x_profiles_MCMC_samples(x,infer.samples)
```

Fortunately, the true value (black line) falls within the posterior 90% credible interval (blue region).

```python
plot_compare_x_profiles_credible_interval(x,KJMA,infer.samples_stats)
```

### Computing the evidence

The evidence $P(d|\beta)$ is estimated using the [Laplace approximation](http://en.wikipedia.org/wiki/Laplace_approximation).

```python
infer.compute_logZ()
print infer.logZ
```

### Grid seach for $v$ and $\ell_0$

In principle, we can also infer the hyperparameters $\beta = [v, \sigma_d, I_0, \sigma_0, \ell_0, \tau_0]$ from the data $d$.

Here, we will infer the fork velocity $v$ and the spatial smoothness scale $\ell_0$ from the data.
If we assume a flat prior for $v$ and $\ell_0$, the posterior probability $P(v,\ell_0|d)$ is simply proportional to the evidence. 

```python
# new infer object from the ki.InferKJMA class to perform the grid search
infer_grid = ki.InferKJMA(x=x, t=t,
                     data=data, xdata=xdata, tdata=tdata,
                     v=v, sigma_d=sigma_d,
                     I_0=I_0, sigma_0=sigma_0, tau_0=tau_0, ell_0=ell_0)
```

```python
# takes about len(vs)*len(ell0_s) min in computation time!
vs = np.arange(1.8,2.2,0.1, dtype=float)
ell_0s = np.arange(10,25,5, dtype=float)
Ps = infer_grid.grid_search_v_ell_0(vs, ell_0s, verbose = True)
grid_search = {'v':vs, 'ell_0':ell_0s, 'P': Ps}
```

The posterior probability for $v$ and $\ell_0$ is peaked at $v=2$ kb/min and $\ell_0=15$ kb.
Fortunately for the fork velocity, the inferred $v$ is equal to the true value.

```python
plot_posterior_v_ell_0(grid_search)
```
