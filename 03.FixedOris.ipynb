### Replication program with well-positioned oris

In some organisms, such as budding yeast, initiation may occur only at pre-defined
sites called **potential origins**. In such cases, the initiation rate reduces to
$I(x,t) = \sum_i \delta(x-x_i) I_i(t)$, where $x_i$ is the position of the potential origin $i$.

We show below an example replication, with two potential oris positioned at $x=30$ kb and $x=60$ kb

```python
v = 2.                                 # replication fork velocity (kb/min)
x = np.arange(100, dtype=float)        # genomic position (kb)
t = np.arange(0, 50, 1/v, dtype=float) # times (min)
x_oris = np.array([30,60])

# artificial initiation rate with a sin^2 I(t) for ori 1, and a gaussian for ori 2 
I_it = np.zeros((2,len(t)))
I_it[0,:] = 0.1*np.sin(2*np.pi*t/40.)**2
I_it[1,:] = 2*np.exp(-0.5*((t - 20)/10.)**2)/np.sqrt(2*np.pi*10.**2)
I_xt = get_I_xt_from_I_it(I_it,x,t,x_oris)
extent=(0,100,0,50)   # (xmin,xmax,tmin,tmax) for images
plot_initiation_rate(I_xt, extent)
```

To give an idea of the resulting stochasticity, we can sample by Monte Carlo simulation five realizations of the replication program, represented by the colored lines below. We celarly see that a potential origin does not initiate at each cell cycle (hence the name potential origin). A potential origin that has not initiated and has been replicated by a fork coming from another origin, is said to be **passively replicated**.

```python
plot_initiation_rate_with_sample(I_xt, extent, x, t, v)
```

### Determining the replication kinetics

The elements of stochastic initiation, constant fork progression, and termination are formally equivalent to the processes of nucleation, growth, and coalescence in crystallization kinetics, and this equivalence has inspired efforts to model DNA replication kinetics using the formalism developed in the 1930s by Kolmogorov, Johnson, Mehl, and Avrami (KJMA) for crystallization kinetics.

Using the KJMA formalism, many quantities describing the replication kinetics can be analytically predicted from the initiation rate $I(x,t)$, for instance:
 
 - $s(x,t)$ the unreplicated fraction (and $f=1-s$ the replicated fraction)
 - $P(x,t)$ the probability distribution of the replication timing at $x$
 - $\rho_{(+)}(x,t)$ the density of right-moving forks
 - $\rho_{(-)}(x,t)$ the density of left-moving forks
 - $\rho_{ini}(x,t)$ the density of initiations
 - $\rho_{ter}(x,t)$ the density of terminations
 - ...
 
 We show below the replication kinetics derived from the initiation rate shown previously.

```python
dx = x[1]-x[0] # spatial resolution
dt = t[1]-t[0] # temporal resolution
lim ={'dy':dx, 'du':dt, 'v':v}

# Let's compute the KJMA kinetics for the artificial initation rate
KJMA = ki.compute_KJMA_kinetics_all(I_xt, lim) 
# and plot
plot_KJMA_kinetics(KJMA, extent)
```

When $I(x,t)=\sum_i \delta(x-x_i) I_i(t)$, the spatial density of initiation reduces to 
$\rho_{ini}(x)=\sum_i \delta(x-x_i) E_i$, where $E_i$, called the **observed efficiency**, is the fraction of cell cycles where the origin $i$ has initiated. 

Due to passive replication, the observed efficiency is not an intrinsic property of an origin: it depends on the firing proprties of the origin *but also* on the firing properties of neighbouring
origins.

```python
plot_observed_efficiency(x, KJMA)
```

### Artificial data

Here, we generate an artificial data set with noise level, spatial and temporal resolutions typical of current experiments.

The experimental unreplicated fraction profiles are usually obtained at a good spatial resolution (around 1 kb), but at a coarser temporal resolution (around 5 min). The true unreplicated fraction
profiles sampled at those resolutions are shown by the colored lines below. 

The artificial data set, shown below, is obtained by adding gaussian noise (of standart deviation $\sigma_d$) to the unreplicated fraction profiles.  

```python
xdata = x.copy()                            # positions
tdata = np.arange(10, 50, 5, dtype=float)   # times
sigma_d = 0.05                              # noise level
true_s, data = ki.generate_artificial_data(xdata,tdata,x,t,I_xt,v,sigma_d)
plot_waterfall_true_s(xdata, tdata, true_s, x, t, KJMA)
plot_waterfall_data(xdata, tdata, data)
```

### Inferring the initiation rate from noisy data

We still adopt a Bayesian, nonparametric approach to infer $I(x,t)$ from the unreplicated fraction data. 

Here, the initiation rate is equal to
$I(x,t) = \sum_i \delta(x-x_i) I_i(t)$, 
where $x_i$ is the position of the potential origin $i$.

Here, we will suppose that we know a priori the positions $x_i$ of the potential origins (as for instance in budding yeast), and we will not try to infer them. For each origin $i$, the initation rate $I_i(t)$ must ne postive and we expect its temporal variation to be smooth. We will assume a Gaussian process on $\log I_i(t)$.

The hyperparameters of our probabilistic model comprise:

- the fork velocity $v$
- the noise level $\sigma_d$
- $\sigma_0$ and $I_0$ set the range of values allowed for the initation rate.
- the temporal smoothness scale $\tau_0$ of the initation rate
- the positions $x_{oris}$ of the potential origins

The mode of the posterior, called the **MAP** (Maximum At Posteriori) estimate, is still obtained by a high-dimensional optimization routine. The MAP estimate of the initation rate, $I_{MAP}(x,t)$, is computed and compared with the true $I(x,t)$ below.

```python
# prior hyperparameters
I_0 = 1e-3
sigma_0 = 3.
ell_0 = 15.
tau_0 = 10.

# All computation will be done in the infer object from the ki.InferKJMA class
infer = ki.InferKJMA(x=x, t=t,
                     data=data, xdata=xdata, tdata=tdata,
                     v=v, sigma_d=sigma_d,
                     I_0=I_0, sigma_0=sigma_0, tau_0=tau_0,
                     FixedOris = True, x_oris = x_oris)

# Let's compute the MAP estimate (takes 1 min in computation time)
infer.compute_MAP(verbose=True)
KJMA_MAP = infer.MAP
```

```python
# Let's compare the true initation rate and the MAP estimate
plot_compare_xt_images(KJMA,KJMA_MAP,key='I',clim=(0,0.06),extent=extent)
```

```python
plot_compare_I_t_profiles(t,I_it,infer.I_MAP)
```

### Very accurate reconstruction of the replication program

We see that $I_{MAP}$ yields a very accurate estimate of the initation rate, exept at the end of S-phase. When $s(x,t)$ is close to zero, it is insensitive even to large variation of $I(x,t)$, hence the initation rate cannot be accurately inferred at the end of S-phase.

Once an estimate of the initation rate is obtained, all other replication kinetics quantities can be analytically predicted, thus completely reconstructing the replication program.

The replication kinetics derived from the MAP estimate $I_{MAP}$ are shown below.

```python
# Let's plot the kinetics derived from the MAP estimate
plot_KJMA_kinetics(KJMA_MAP, extent)
```

The reconstructed quantities are almost indistinguishable from their true values.
Furthermore, note that all those quantities are reconstructed at a small temporal resolution of 0.5 min, while the original data (see above) have only a temporal resolution of 5 min! This interpolation is indeed possible thanks to the temporal smoothness scale $\tau_0$= 10 min.

Below, we compare the estimates obtained for the mean replication timing $T(x)$, the replication fork polarity $p(x)$, the initiation density $\rho_{ini}(x)$ and the termination density $\rho_{ter}(x)$  (red dashed lines) with their true values (green plain lines).

```python
# quantitative comparision for the x-profiles
plot_compare_x_profiles_wellpos_oris(x,KJMA,KJMA_MAP)
```

### MCMC sampling

The MAP estimate yields a very accurate reconstruction of the replication program, however it does not give information on the reliability (or posterior uncertaincy) of the estimate.

In order to estimate the posterior uncertainty, we sample the posterior distribution $P(I|d, \beta)$ using a MCMC method called the [Hamiltonian Monte Carlo algorithm](http://en.wikipedia.org/wiki/Hamiltonian_Monte_Carlo).

```python
infer.compute_MCMC_samples(verbose=True)
```

```python
plot_x_profiles_MCMC_samples_wellpos_oris(x,infer.samples)
```

```python
plot_compare_x_profiles_credible_interval_wellpos_oris(x,KJMA,infer.samples_stats)
```

```python
fig, axs = plt.subplots(2,2,sharex=True, sharey=True, figsize=(8,8))
for (i,j), ax in np.ndenumerate(axs):
    key = 'I'
    xj = x_oris[j]
    if i==0:
        for s in [10,20,30,40,50]: ax.plot(t,infer.samples[key][xj,:,s],color='blue',alpha=0.3)
    if i==1:
        y1 = infer.samples_stats['5'][key][xj,:]
        y2 = infer.samples_stats['95'][key][xj,:]
        ax.fill_between(t,y1,y2,color='blue',alpha=0.3)
        ax.plot(t,KJMA[key][xj,:],color='k', alpha=0.5, linewidth=2)
    remove_junk_xyplot(ax)
    ax.set_ylabel('I(x,t)')
    ax.set_xlabel('t (min)')
    ax.locator_params(nbins=3)
    ax.set_ylim(0,0.2)
fig.tight_layout()
```
