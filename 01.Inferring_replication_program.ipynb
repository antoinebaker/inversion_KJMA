### Introduction

Cells must accurately duplicate their DNA content at every cell division, during the synthesis phase (**S-phase**) of the cell cycle. DNA replication initiates at genomic sites called **origins of replication**. The DNA is copied by a pair of oppositely moving **replication forks** that propagate away from each origin. These forks actively copy the genome away from the origin until they encounter another replication fork at a **replication terminus**.

<img src="dna-replication-2.png" />

The spatio-temporal organization of the replication program (where along the genome and when during the S-phase replication initiates) is of fundamental importance to the cellular life. When the cell fails to accurately and timely replicate its DNA, it leads to cell death or cancer.  

Recent experimental progress allows to determine the fraction of replicated DNA along the genome and at different times during the S-phase. Is it possible, from the **replicated fraction data**, to infer the underlying replication program ?

### Stochastic replication program

The replication program is inherently **stochastic**: the number of origins used, their positions along the genome, and their firing times all change from one cell cycle to another. A stochastic replication program is completely specified by the **initiation rate $I(x,t)$**, defined as the probability of the locus $x$ to initiate at time $t$.  

An example replication program, with two "blobs" of initiations is shown below.

```python
v = 2.                                 # replication fork velocity (kb/min)
x = np.arange(100, dtype=float)        # genomic position (kb)
t = np.arange(0, 50, 1/v, dtype=float) # times (min)

# artificial initiation rate constituted of two "gaussian blobs"
I_xt = 5*ki.gaussian_blob(x,t,30,20,5,3) + 5*ki.gaussian_blob(x,t,60,20,5,3)
extent=(0,100,0,50)   # (xmin,xmax,tmin,tmax) for images
plot_initiation_rate(I_xt, extent)
```

To give an idea of the resulting stochasticity, we can sample by Monte Carlo simulation three realizations of the replication program, represented by the colored lines below.

```python
plot_initiation_rate_with_sample(I_xt, extent, x, t, v)
```

### Determining the replication kinetics

The elements of stochastic initiation, constant fork progression, and termination are formally equivalent to the processes of nucleation, growth, and coalescence in crystallization kinetics, and this equivalence has inspired efforts to model DNA replication kinetics using the formalism developed in the 1930s by Kolmogorov, Johnson, Mehl, and Avrami (KJMA) for crystallization kinetics.

Using the KJMA formalism, many quantities describing the replication kinetics can be analytically predicted from the initiation rate $I(x,t)$, for instance:
 
 - $s(x,t)$ the unreplicated fraction (and $f=1-s$ the replicated fraction)
 - $P(x,t)$ the probability distribution of the replication timing at $x$
 - $\rho_{(+)}(x,t)$ the density of right-moving forks
 - $\rho_{(-)}(x,t)$ the density of left-moving forks
 - $\rho_{ini}(x,t)$ the density of initiations
 - $\rho_{ter}(x,t)$ the density of terminations
 - ...
 
 We show below the replication kinetics derived from the initiation rate shown previously.

```python
dx = x[1]-x[0] # spatial resolution
dt = t[1]-t[0] # temporal resolution
lim ={'dy':dx, 'du':dt, 'v':v}

# Let's compute the KJMA kinetics for the artificial initation rate
KJMA = ki.compute_KJMA_kinetics_all(I_xt, lim) 
# and plot
plot_KJMA_kinetics(KJMA, extent)
```

### Inferring the initiation rate from noisy data

Here, we will adopt a Bayesian, nonparametric approach to properly infer $I(x,t)$ from the unreplicated fraction data. 

In a **Bayesian approach**, all types of uncertainty (*eg* on the parameters, on the model to use, due to intrinsic noise or due to incomplete knowledge) are consistently represented by probabilities. Thus, Bayesian methods are also called probabilistic methods.
The Bayesian formulation is particularly well adapted to parameter estimation problems. In our case, the goal is to infer the parameter $I$ (the initiation rate) from the data $d$ (a noisy measurement of the unreplicated fraction). We recall that the	posterior probability of $I$, given data $d$, is determined	by **Bayes’ theorem**:

$\underbrace{P(I|d \beta)}_{posterior} = \underbrace{P(d|I\beta)}_{likelihood} \underbrace{P(I|\beta)}_{prior} / \underbrace{P(d| \beta)}_{evidence}$

The likelihood follows the noise model for the data, while the prior encodes any available information—even vague—about the parameter to infer. 

The inference procedure is technically challenging because:

- the "parameter" to infer, the initiation rate $I(x,t)$, is a function of space and time. In other words, we have to infer an infinite-dimensional parameter. Actually, we want to infer $I$ at a finite resolution, for instance 1 kb and space and 0.5 min in time, but then $I(x,t)$ is still a very high ($10^4$) dimensional vector. A model that comprises an infinite or very large number of parameters is somehow paradoxically called a **non-parametric** model.
- $I(x,t)$ is a rate, and must be positive. Besides, we expect that temporal and spatial variations of $I(x,t)$ are smooth. We enforce these constraints by assuming a [Gaussian process](http://en.wikipedia.org/wiki/Gaussian_process) prior on the log initiation rate.
- the relationship between the data $d$ and $I$ is highly non-linear. Therefore the analytical formulas used in classic Gaussian process regression tasks cannot be used.
 
The specification of a probabilistic model for the likelihood and the prior requires an additional set of parameters, called hyperparameters, symbolized by $\beta$ in the Bayes' theorem. In our case, the hyperparameters comprise:

- the fork velocity $v$
- the noise level $\sigma_d$
- $\sigma_0$ and $I_0$ set the range of values allowed for the initation rate.
- the spatial smoothness scale $\ell_0$ of the initation rate
- the temporal smoothness scale $\tau_0$ of the initation rate
 
The mode of the posterior, called the **MAP** (Maximum At Posteriori) estimate, can be obatined by a high-dimensional optimization routine. The MAP estimate of the initation rate, $I_{MAP}(x,t)$, is computed and compared with the true $I(x,t)$ below.

```python
# prior hyperparameters
I_0 = 1e-4
sigma_0 = 3.
ell_0 = 15.
tau_0 = 10.

# All computation will be done in the infer object from the ki.InferKJMA class
infer = ki.InferKJMA(x=x, t=t,
                     data=data, xdata=xdata, tdata=tdata,
                     v=v, sigma_d=sigma_d,
                     I_0=I_0, sigma_0=sigma_0, ell_0=ell_0, tau_0=tau_0)

# Let's compute the MAP estimate (takes 1 min in computation time)
infer.compute_MAP(verbose=True)
KJMA_MAP = infer.MAP
```

```python
# Let's compare the true initation rate and the MAP estimate
plot_compare_xt_images(KJMA,KJMA_MAP,key='I',clim=(0,0.06),extent=extent)
```

### Very accurate reconstruction of the replication program

We see that $I_{MAP}$ yields a very accurate estimate of the initation rate.
But once an estimate of the initation rate is obtained, all other replication kinetics quantities can be analytically predicted, thus completely reconstructing the replication program.

The replication kinetics derived from the MAP estimate $I_{MAP}$ are shown below.

```python
# Let's plot the kinetics derived from the MAP estimate
plot_KJMA_kinetics(KJMA_MAP, extent)
```

The reconstructed quantities are almost indistinguishable from their true values.
Furthermore, note that all those quantities are reconstructed at a small temporal resolution of 0.5 min, while the original data (see above) have only a temporal resolution of 5 min! This interpolation is indeed possible thanks to the temporal smoothness scale $\tau_0$= 10 min.

Below, we compare the estimates obtained for the mean replication timing $T(x)$, the replication fork polarity $p(x)$, the initiation density $\rho_{ini}(x)$ and the termination desnity $\rho_{ter}(x)$  (red dashed lines) with their true values (green plain lines).

```python
# quantitative comparision for the x-profiles
plot_compare_x_profiles(x,KJMA,KJMA_MAP)
```

Finally, we illustrate below the accurate interpolation achieved by the inference procedure.
From noisy data, obtained at a low temporal resolution of 5 min, we are able to accurately infer
the unreplicated fraction at a 0.5 min resolution (MAP esimate in red, true values in gray).
